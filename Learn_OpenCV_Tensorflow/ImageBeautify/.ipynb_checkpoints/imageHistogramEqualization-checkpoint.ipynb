```python
# 灰度图片的直方图均衡化
import cv2
import numpy as np
img = cv2.imread('thimg.jpg',1)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('src',gray)
dst = cv2.equalizeHist(gray)
cv2.imshow('dst',dst)
cv2.waitKey(0)
```

```python
# 彩色图片的直方图均衡化
import cv2
import numpy as np
img = cv2.imread('thimg.jpg',1)
cv2.imshow('src',img)
(b,g,r) = cv2.split(img)   #通道分解
bH = cv2.equalizeHist(b)
gH = cv2.equalizeHist(g)
rH = cv2.equalizeHist(r)
result = cv2.merge((bH,gH,rH))   #通道合成
cv2.imshow('dst',result)
cv2.waitKey(0)
```

```python
# YUV的直方图均衡化
import cv2
import numpy as np
img = cv2.imread('thimg.jpg',1)
#imgYUV = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
imgYUV = cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
#imgYUV = cv2.cvtColor(img,cv2.COLOR_BGR2YUV_I420)
#imgYUV = cv2.cvtColor(img,cv2.COLOR_BGR2YUV_IYUV)
#imgYUV = cv2.cvtColor(img,cv2.COLOR_BGR2YUV_YV12)
cv2.imshow('src',img)
channelYUV = cv2.split(imgYUV)
channelYUV[0] = cv2.equalizeHist(channelYUV[0])
channels = cv2.merge(channelYUV)
#result = cv2.cvtColor(channels,cv2.COLOR_YCrCb2BGR)
result = cv2.cvtColor(channels,cv2.COLOR_YUV2BGR)
cv2.imshow('dst',result)
cv2.waitKey(0)
```

```python
#gray pic Hist 均衡化
#累积概率：
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('thimg.jpg',1)
cv2.imshow('src',img)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
count = np.zeros(256,np.float)
for i in range(0,height):
    for j in range(0,width):
        pixel = gray[i,j]
        index = int(pixel)
        count[index] = count[index]+1
for i in range(0,256):
    count[i] = count[i]/(height*width)
#计算累积概率
sum1 = float(0)
for i in range(0,256):
    sum1= sum1+ count[i]
    count[i] = sum1
#print(count)
#计算映射表
map1 = np.zeros(256,np.uint16)
for i in range(0,256):
    map1[i] = np.uint16(count[i]*255)
#映射
for i in range(0,height):
    for j in range(0,width):
        pixel = gray[i,j]
        gray[i,j] = map1[pixel]
cv2.imshow('dst',gray)
cv2.waitKey(0)
```

```python
#彩色直方图 均衡化
#
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('thimg.jpg',1)
cv2.imshow('src',img)
imgInfo = img.shape
height = imgInfo[0]
width = imgInfo[1]
count_b = np.zeros(256,np.float)
count_g = np.zeros(256,np.float)
count_r = np.zeros(256,np.float)
for i in range(0,height):
    for j in range(0,width):
        (b,g,r) = img[i,j]
        index_b = int(b)
        index_g = int(g)
        index_r = int(r)
        count_b[index_b] = count_b[index_b]+1
        count_g[index_g] = count_g[index_g]+1
        count_r[index_r] = count_r[index_r]+1
for i in range(0,256):
    count_b[i] = count_b[i]/(height*width)
    count_g[i] = count_g[i]/(height*width)
    count_r[i] = count_r[i]/(height*width)
#计算累积概率
sumb = float(0)
sumg = float(0)
sumr = float(0)
for i in range(0,256):
    sumb= sumb+ count_b[i]
    count_b[i] = sumb
    sumg= sumg+ count_g[i]
    count_g[i] = sumg
    sumr= sumr+ count_r[i]
    count_r[i] = sumr
#print(count)
#计算映射表
mapb = np.zeros(256,np.uint16)
mapg = np.zeros(256,np.uint16)
mapr = np.zeros(256,np.uint16)
for i in range(0,256):
    mapb[i] = np.uint16(count_b[i]*255) 
    mapg[i] = np.uint16(count_g[i]*255)
    mapr[i] = np.uint16(count_r[i]*255)
#映射
dst = np.zeros((height,width,3),np.uint8)
for i in range(0,height):
    for j in range(0,width):
        (b,g,r) = img[i,j]
        b = mapb[b]
        g = mapg[g]
        r = mapr[r]
        dst[i,j] = (b,g,r)
cv2.imshow('dst',dst)
cv2.waitKey(0)
```
